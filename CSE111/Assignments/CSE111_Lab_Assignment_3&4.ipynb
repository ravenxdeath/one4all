```python
#Task2

class Shape:
    def __init__(self,name,side1,side2):
        self.name = name
        self.side1 = side1          
        self.side2 = side2
        
    def area(self):
            if self.name == "Triangle" or self.name=="Rhombus":
                area = .5*self.side1*self.side2
            
            elif self.name == "Square" or self.name=="Rectangle":
                area = self.side1*self.side2
            else:
                area = "Shape Unknown"
            
            print(f"Area: {area} ")

# Write your code here
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
# Task 3

class Calculator:
    def __init__(self):
        print("Calculator is ready!")
        
    def calculate(self,v1,v2,sign):
        self.v1 = v1
        self.v2 = v2
        self.sign = sign
        
        if sign == "+":
            self.ret = v1 + v2
        elif sign == "-":
            self.ret = v1 - v2
        elif sign == "*":
            self.ret = v1 * v2
        elif sign == "/":
            self.ret = v1 / v2
        return self.ret
    
    def showCalculation(self):
        print(self.v1,self.sign, self.v2,"=", self.ret)
        
        

# Write your code here
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#task 13
class StudentDatabase:
  def __init__(self,name,identity):
    self.name=name
    self.id=identity
    self.grades={}
    
  def calculateGPA(self,lst, season):
    self.sub=[]
    self.cg=0
    for i in lst:
      ko=i.split(':')
      self.sub.append(ko[0])
      x=float(ko[1])
      self.cg+=x
    self.gpa= self.cg/len(self.sub)
    self.gpa= round(self.gpa,2)
    sub=tuple(self.sub)
    self.dic2={}
    self.dic2[sub]=self.gpa
    self.grades[season]= self.dic2
    
  def printDetails(self):
    print(f'Name: {self.name} \nID: {self.id}')
    print(f"Courses taken in  ")
    for i,j in self.grades.items():
      print(f'Courses taken in {i}')
      for k,l in j.items():
        for m in k:
          print(m)
        print(f'GPA: {l}')

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
#ungraded -23
class Author:
  def __init__(self, n=None):
    self.name=n
    self.dic={}
    self.bookcount=0
  
  def addBook(self,bn,bg):
    if self.name != None:
      self.bookname=bn
      self.bookgenre=bg
      if bg not in self.dic:
        self.dic[bg]=[bn]
      else:
        self.dic[bg].append(bn)
      self.bookcount+=1
    else:
      print('A book can not be added without author name')

  def setName(self,x):
    self.name=x
  def printDetail(self):
    print(f'Number of Book(s) : {self.bookcount}')
    print(f'Author Name: {self.name}')
    for i,j in self.dic.items():
      x=f'{j}'
      print(f'{i} : {x[1:-1]}')

a1 = Author()
print("=================================")
a1.addBook('“Ice”', 'Science Fiction')
print("=================================")
a1.setName('Anna Kavan')
a1.addBook('“Ice”', 'Science Fiction')
a1.printDetail()
print("=================================")
a2 = Author('Humayun Ahmed')
a2.addBook('“Onnobhubon”', 'Science Fiction')
a2.addBook('Megher Upor Bari', '“Horror”')
print( "=================================")
a2.printDetail()
a2.addBook('“Ireena”', 'Science Fiction')
print("=================================")
a2.printDetail()
print("=================================")
print(a2.dic)
 
```

```python
#ungraded - 24
class Hospital:
  def __init__(self,n):
    self.name=n
    self.doclist={}
    self.patlist={}

  def addDoctor(self,doc):
    self.doclist[doc.id]=[doc.name , doc.speciality]
  def addPatient(self,pat):
    self.patlist[pat.id]=[pat.name, pat.age, pat.phn]

  def getDoctorByID(self,id):
    do= f'Doctor ID: {id}'
    na= f'Name : {self.doclist[id][0]}'
    sp= f'Speciality :{self.doclist[id][1]}'
    return f'{do} \n{na} \n{sp}'
    
  def getPatientByID(self,id):
    do= f'Patients ID: {id}'
    na= f'Name : {self.patlist[id][0]}'
    ag= f'Age :{self.patlist[id][1]}'
    ph= f'Phone No: {self.patlist[id][2]}'
    return f'{do} \n{na} \n{ag} \n{ph}'

  def allDoctors(self):
    print('All Doctors:')
    print(f'Number of Doctors: {len(self.doclist)}')
    print(f'{self.doclist}')
  def allPatients(self):
    print('All Patients: ')
    print(f'Number of patients: {len(self.patlist)}')
    print(f'{self.patlist}')


class Doctor:
  def __init__(self,i,doc,name,spe):
    self.id=i
    self.profession=doc
    self.name=name
    self.speciality=spe

class Patient:
  def __init__(self,i,pat,name,age,phn):
    self.id=i
    self.name=name
    self.age=age
    self.phn=phn

h = Hospital("Evercare")
d1 = Doctor("1d","Doctor", "Samar Kumar", "Neurologist")
h.addDoctor(d1)
print("=================================")
print(h.getDoctorByID("1d"))
print("=================================")
p1 = Patient("1p","Patient", "Kashem Ahmed", 35, 12345)
h.addPatient(p1)
print("=================================")
print(h.getPatientByID("1p"))
print("=================================")
p2 = Patient ("2p","Patient", "Tanina Haque", 26, 33456)
h.addPatient(p2)
print("=================================")
print(h.getPatientByID("2p"))
print("=================================")
h.allDoctors()
h.allPatients()
```

```python
#ungraded-25
class Vaccine:
  def __init__(self,n,c,d):
    self.vacname=n
    self.country=c
    self.days=d

class Person:
  def __init__(self,n,a,t='General Citizen'):
    self.name=n
    self.age=a
    self.types=t
    self.vac=''
    self.dose=0

  def pushVaccine(self,vaccine,d=None):
    if self.age<25 and  self.types != 'Student':
      print(f'Sorry {self.name}.Minimum age is 25years now')
    else:
      if d ==None:
        print(f' 1st dose done for {self.name}')
        self.vaccinename = vaccine.vacname
        self.gapdays= vaccine.days
        self.vac= vaccine.vacname
      else:
        self.dose=2
        if self.vac == vaccine.vacname:
          print(f'{d[0]}nd dose done for {self.name}')
        else:
          print(f'Sorry {self.name}, you cant take 2 diff vaccines.')

  def showDetail(self):
    print(f'Name: {self.name} Age:{self.age} Type: {self.types}')
    print(f'Vaccine Name: {self.vaccinename}')
    print(f'1st dose : Given')
    if self.dose ==2:
      print(f'2nd dose : Given')
    else:
      print(f'2nd dose: Please come after {self.gapdays} days')


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
