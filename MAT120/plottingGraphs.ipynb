```python
from sympy import*

x = symbols("x")

f = 3*x**4 + 2*x**3 + 5*x**2+ 2*x + 3

df = diff(f, x)
df2 = diff(f, x, 2)
df3 = diff(f, x, 3)

print(f, df, df2, df3, sep=" || ")

f = lambdify(x, f)
df = lambdify(x, df)
df2 = lambdify(x, df2)
df3 = lambdify(x, df3)

import numpy as np

p = np.arange(-7,6)
a = f(p)

q = np.arange(-5,6)
b = df(q)

r = np.arange(-8,8)
c = df2(r)

s = np.arange(-9,2)
d = df3(s)

print(f"p:{p}\nq:{q}\nr:{r}\ns:{s}")
print(f'{a}\n{b}\n{c}\n{d}')

import matplotlib.pyplot as plt
plt.plot(p,a)
plt.xlabel("x axis")    
plt.ylabel("y axis")
plt.title("Shape of Equation")
plt.show()
```
