```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial
def f(x):
    return x**5 + 2.5*x**4 - 2*x**3 - 6*x**2 + x/2 + 2

g2_x=((1/6)*(x**5 + 2.5*x**4 - 2*x**3 + x/2 + 2))**1/2
g3_x=((1/2)*(x**5 + 2.5*x**4 - 6*x**2 + x/2 + 2))**1/3

def g1(x):
  p = Polynomial([-2.0, 0.0, 6.0, 2.0, -2.5, -1.0])
  return (p(x)/2)

def g2(x):
  p = Polynomial([2.0, 0.5, 0.0, -2.0, 2.5, 1.0])
  return np.sqrt(p(x)/6)

def g3(x):
  p = Polynomial([2.0, 0.5, -6.0, 0.0, 2.5, 1.0])
  return np.cbrt(p(x)/2)
```
